```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
## for Palatino and other serif fonts use:
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

```python
nodes=[1,2,4,8]
acc=[1.000,2.022,4.026,7.823]
```

```python
def adjust(x):
    y=8*x/6-1/3
    return y
```

```python
xline=np.arange(0,9,0.01)
yline=adjust(xline)
```

```python
plt.plot(nodes,acc,'.',label="Performance")
plt.plot(xline,yline,label="$y=1.33*x-0.33$")
plt.plot(xline,xline,label="$y=x$")
plt.ylabel("Speed-up")
plt.xlabel("Num of Nodes")
plt.xlim(1,9)
plt.ylim(1,9)
plt.legend()
plt.savefig("acc.png",dpi=500)
```
